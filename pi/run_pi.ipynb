```python
#run on all spar strains

results = {}

directory = "/Users/clairedubin/spur/peroxisomes/alignments/scer_spar_by_scer_population_091919/1WineEuropean/"

all_files = os.listdir(directory)  

for i, file in enumerate(all_files):
    

    start_time = time.time()

    print(i, file)

    gene = file.split('_')[3]

    seq_dict = []

    for record in SeqIO.parse(directory+file, "fasta"):

        if record.id == 'Spar;':

            if 'N-44' not in record.description and 'YPS138' not in record.description:

                seq_dict += [(record.description, record.seq)]   

    if len(seq_dict) > 8:

        result = pi(seq_dict)
        print(gene, result)
        results[gene] = result

        with open('/Users/clairedubin/spur/peroxisomes/pi/pi_eurospar_032620.csv', 'a') as f:
            
            writer = csv.writer(f)
            writer.writerow([gene, result])

            f.close()

    end_time = time.time()

    print("total time taken this loop: ", end_time - start_time)
```
